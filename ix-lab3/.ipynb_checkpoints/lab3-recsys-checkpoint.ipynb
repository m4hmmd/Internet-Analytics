###  The user biases {Î±u} and the item biases {Î²m}

```python
def cumulate_by_key(rdd):
    return rdd.reduceByKey(lambda x, y: x + y).sortByKey()
```

```python
def count_by_key_rdd(rdd):
    dict_rdd_count = rdd.countByKey().items()
    return sc.parallelize(dict_rdd_count)
```

```python
def change_list_dict(lst):
    dic = {}
    for i in lst:
        dic[i[0]] = i[1]
    return dic
```

```python
def key_bias(rdd):
    # Cumulate the rating values for each key
    cum_rdd = cumulate_by_key(rdd)

    # Count the number of ratings for each key
    count_rdd = count_by_key_rdd(rdd)

    # Join the two sets by their keys
    joined_rdd = cum_rdd.leftOuterJoin(count_rdd).sortByKey()
    
    # Calculate the user bias for each user
    lst_key_bias = joined_rdd.map(lambda a: (a[0], a[1][0] / a[1][1])).collect()
 
    # Finally, change the list to a dictionary
    dict_key_bias = change_list_dict(lst_key_bias)

    return dict_key_bias
```

```python
# Minus Âµ from all of the rating values
user_rating_Âµ = training.map(lambda r: (r[0], r[2] - Âµ))

# Minus Âµ and user bias from all of the rating values
movie_rating_Âµ = training.map(lambda r: (r[1], r[2] - user_bias[r[0]] - Âµ))
```

```python
user_bias = key_bias(user_rating_Âµ)
movie_bias = key_bias(movie_rating_Âµ)
```

###  Ex 3.7 Evaluation

Implement a function error() that takes an RDD containing (userId, movieId, rating) triplets and computes the error (3.2) with respect to the validation set.
- Use this function to evaluate the baseline predictions that you computed previously.
- You should obtain an error of approximately 0.867.

```python
# Calculates an error
# used for evaluation of the recommender systems
def error(predictions, validation):
   # Attach keys to values in the prediction and validation sets
    predictions_by_user_product = predictions.map(lambda r: ((r[0], r[1]), r[2]))
    validations_by_user_product = validation.map(lambda r: ((r[0], r[1]), r[2]))

    # Join the two sets by their keys (user, product)
    joined_by_user_product = predictions_by_user_product.join(validations_by_user_product)

    # Calculate the square of the differences for each pair of validation and prediction
    # and re-key by user
    sq_differences_by_user = (joined_by_user_product
        .map(lambda x: (x[0][0], (x[1][0] - x[1][1])**2)))

    # Take the squareroot of the mean of the square differences from above
    # on a user-by-user basis to get each user's error
    user_errors = (sq_differences_by_user
        .map(lambda x: (x[0], (x[1], 1)))
        .reduceByKey(lambda a, b: (a[0] + b[0], a[1] + b[1]))
        .map(lambda x: np.sqrt(x[1][0] / x[1][1])))

    # Finally, take the mean of the each user's error
    mean_error = user_errors.mean()

    return mean_error
```

## 3.2.2 Matrix-Factorization Model

```python
from pyspark.mllib.recommendation import *
```

```python
# Load the personalized data from file
with open('my-ratings.txt') as personal_ratings_file:
    data1 = personal_ratings_file.read().splitlines()
    personal_data = sc.parallelize(data1).map(json.loads)
# Union the original data and the personal data
unioned_data = data.union(personal_data)
```

```python
VALIDATION_SPLIT_RATIO = 0.2
```

```python
# First split data into test and training data
sec_322_training_raw = unioned_data.filter(lambda x: x['timestamp'] % 10 >= 10*VALIDATION_SPLIT_RATIO)
sec_322_validation_raw = unioned_data.filter(lambda x: x['timestamp'] % 10 < 10*VALIDATION_SPLIT_RATIO)

# Then marshall to the correct format
sec_322_training_ratings = sec_322_training_raw.map(lambda x: Rating(x['userId'], x['movieId'], x['rating']))
sec_322_validation_ratings = sec_322_validation_raw.map(lambda x: Rating(x['userId'], x['movieId'], x['rating']))
sec_322_validation_x = sec_322_validation_ratings.map(lambda r: (r.user, r.product))
sec_322_validation_y = sec_322_validation_ratings.map(lambda r: r.rating)
```

```python
# Calculates an error
# used for evaluation of the recommender systems
def error(predictions, validation):
    # Attach keys to values in the prediction and validation sets
    predictions_by_user_product = predictions.map(lambda r: ((r.user, r.product), r.rating))
    validations_by_user_product = validation.map(lambda r: ((r.user, r.product), r.rating))
    
    # Join the two sets by their keys (user, product)
    joined_by_user_product = predictions_by_user_product.join(validations_by_user_product)
    
    # Calculate the square of the differences for each pair of validation and prediction
    # and re-key by user
    sq_differences_by_user = (joined_by_user_product
        .map(lambda x: (x[0][0], (x[1][0] - x[1][1])**2)))
    
    # Take the squareroot of the mean of the square differences from above
    # on a user-by-user basis to get each user's error
    user_errors = (sq_differences_by_user
        .map(lambda x: (x[0], (x[1], 1)))
        .reduceByKey(lambda a, b: (a[0] + b[0], a[1] + b[1]))
        .map(lambda x: np.sqrt(x[1][0] / x[1][1])))
    
    # Finally, take the mean of the each user's error
    mean_error = user_errors.mean()
    
    return mean_error
```

### Exercise 3.8 Regularization

We will start by experimenting with the rank of the matrix factorization model.

```python
# CAUTION: Run this cell only to clear the saved trained models.
# Initialize a dictionary of possible models, referenced by rank of the matrix factorization.
sec_322_model_by_rank = {}
```

```python
# CAUTION: Run this cell only to clear the saved predictions.
# Initialize a dictionary of predictions, referenced by rank of the model used to generate them.
sec_322_prediction_by_rank = {}
```

```python
# CAUTION: Run this cell only to clear the saved error values.
# Initialize a dictionary of validation error values, referenced by rank of the model used to generate them.
sec_322_validation_error_by_rank = {}
```

```python
# The ranks to try out
sec_322_ranks = [5, 10, 15, 20, 25]
```

```python
# Train models for each of the ranks to try out
for rank in sec_322_ranks:
    print()
    print('====== rank {} ======'.format(rank))
    if rank not in sec_322_model_by_rank:
        print('training...')
        model = ALS.train(sec_322_training_ratings, rank)
        print('done training.')
        sec_322_model_by_rank[rank] = model
        print('trained model saved.')
    else:
        model = sec_322_model_by_rank[rank]
        print('trained model loaded.')
    if rank not in sec_322_prediction_by_rank:
        print('predicting...')
        predicted_ratings = model.predictAll(sec_322_validation_x)
        print('done predicting.')
        sec_322_prediction_by_rank[rank] = predicted_ratings
        print('predictions saved.')
    else:
        predicted_ratings = sec_322_prediction_by_rank[rank]
        print('predictions loaded.')
    if rank not in sec_322_validation_error_by_rank:
        print('calculating error between predictions and validation set...')
        validation_error = error(predicted_ratings, sec_322_validation_ratings)
        print('validation error calculated.')
        sec_322_validation_error_by_rank[rank] = validation_error
        print('validation error saved.')
    else:
        validation_error = sec_322_validation_error_by_rank[rank]
        print('validation error loaded.')
    print('for rank = {}, the error is {:.4f}'.format(rank, validation_error))
```

**Trade-off between higher vs lower rank:** It seems like as we increase the rank of the matrix factorization, we lose accuracy in our validation. The benefit of having a higher rank is a more flexible model, but this extra flexibility may be misused through overfitting.

```python
# Pick rank of 10 since it produces a relatively small error, and doesn't sacrifice 
# as much flexibility as the rank-5 model
sec_322_RANK = 10
```

```python
# Helper to train with various lambdas
def train_and_evaluate(lambda__,
                       training_ratings=sec_322_training_ratings,
                       rank=sec_322_RANK,
                       validation_ratings=sec_322_validation_ratings,
                       validation_x=sec_322_validation_x):
    print('training model...')
    model = ALS.train(training_ratings, rank, lambda_=lambda__)
    print('model trained.')
    print('validating...')
    predictions = model.predictAll(validation_x)
    print('predictions complete.')
    print('calculating validation error...')
    validation_error = error(predictions, validation_ratings)
    print('validation error calculated.')
    return model, predictions, validation_error
```

```python
# Use this value as the lambda (regularization parameter)
sec_322_LAMBDA_TINY = 1.0e-4
```

```python
# Train using the selected rank with the selected regularization factor
_, _, sec_322_lambda_tiny_error = train_and_evaluate(sec_322_LAMBDA_TINY)
```

```python
# Compare to the original one (lambda_ was 1.0e-2 by default)
print('for lambda = {:.4f}, the error is {:.4f}'.format(sec_322_LAMBDA_TINY, sec_322_lambda_tiny_error))
print('for lambda = {:.4f}, the error is {:.4f}'.format(1.0e-2, sec_322_validation_error_by_rank[sec_322_RANK]))
```

**Comparison of $\lambda_{tiny} = 10^{-4}$ vs $\lambda_{default} = 10^{-2}$:**

The model using a *lower* regularization parameter of just $10^{-4}$ instead of the original $10^{-2}$ results in a higher validation error. This is because removing regularization increases the model's effective degrees of freedom. This increase in flexibility in turn causes the overfitting of the model to the training data, making it less accurate with the "new" validation data because it is "fit to the noise".

```python
# Use this vale as the lambda now
sec_322_LAMBDA_HUGE = 1.0e1
```

```python
# Train using the selected rank with the selected regularization factor
_, _, sec_322_lambda_huge_error = train_and_evaluate(sec_322_LAMBDA_HUGE)
```

```python
# Compare all three lambdas
print('  lambda |  error')
print('{:8.4f} | {:6.4f}'.format(sec_322_LAMBDA_HUGE, sec_322_lambda_huge_error))
print('{:8.4f} | {:6.4f}'.format(1.0e-2, sec_322_validation_error_by_rank[sec_322_RANK]))
print('{:8.4f} | {:6.4f}'.format(sec_322_LAMBDA_TINY, sec_322_lambda_tiny_error))
```

**Comparison with $\lambda = 10.0$:**

The model using a *higher* regularization parameter of a whopping $10.0$ rather than the original $10^{-2}$ also results in a much, much higher validation error. This is because overregularizing the model significantly decreases the model's effective degrees of freedom. This lack of flexibility means that the model will be overly generalized and cannot capture the intricacies of details in the data, resulting in the lower validation accuracy.

In the case of a $\lambda$ as large as $10.0$, the regularizer term becomes the dominant term in much of the training loss function, meaning that training itself converges slower and is less effective.

```python
# Initialize a list of lambda values, start with the default lambda and the tiny one
sec_322_lambdas = [sec_322_LAMBDA_TINY, 0.01]
```

```python
# Initialize a list of validation errors (with starting values corresponding to starting lambda values)
sec_322_validation_errors = [sec_322_lambda_tiny_error,
                             sec_322_validation_error_by_rank[sec_322_RANK]]
```

```python
# Procedure to iterate on lambda values
def iterate_on_lambda(iterations, initial_step_factor, diminish_rate):
    for _ in range(iterations):
        iteration = len(sec_322_lambdas) - 2
        print()
        print('===== Iteration {} ====='.format(iteration))
        print()
        print('lambda Error gradient:')
        print('{:5f} - {:5f}'.format(sec_322_validation_errors[-1], sec_322_validation_errors[-2]))
        print('---------------------')
        print('{:5f} - {:5f}'.format(sec_322_lambdas[-1], sec_322_lambdas[-2]))
        grad = (sec_322_validation_errors[-1] - sec_322_validation_errors[-2]) / (sec_322_lambdas[-1] - sec_322_lambdas[-2])
        print('= {:5f}'.format(grad))
        # Like a learning rate
        sec_322_LAMBDA_step_factor = initial_step_factor * diminish_rate**iteration
        print()
        print('step factor is {:5f}'.format(sec_322_LAMBDA_step_factor))
        delta_lambda = -np.sign(grad) * sec_322_LAMBDA_step_factor
        print()
        print('lambda will be adjusted by {:5f}'.format(delta_lambda))
        new_lambda = sec_322_lambdas[-1] + delta_lambda
        print()
        print('old lambda = {:5f}; new lambda = {:5f}'.format(sec_322_lambdas[-1], new_lambda))
        print()
        print('training and evaluating new model...')
        _, _, validation_error = train_and_evaluate(new_lambda)
        print('trainging and evaluation complete.')
        print()
        print('old validation error = {:5f}; new validation error = {:5f}'.format(sec_322_validation_errors[-1], validation_error))
        print('with an improvement of {:5f}'.format(sec_322_validation_errors[-1] - validation_error))
        sec_322_lambdas.append(new_lambda)
        sec_322_validation_errors.append(validation_error)

# Procedure to go back to a previous lambda value and discard future iterations
def go_back_to_iteration(n):
    sec_322_lambdas = sec_322_lambdas[:n]
    sec_322_validation_errors = sec_322_validation_errors[:n]
```

```python
# Rerun this block to iterate more and try the next lambda value
iterate_on_lambda(10, initial_step_factor=0.01, diminish_rate=0.95)
```

```python
# Summary of lambdas and their validation errors
print('iteration |   lambda |    error')
for i, (lambda_, val_error) in enumerate(zip(sec_322_lambdas, sec_322_validation_errors)):
    print('{:9d} | {:5f} | {:5f}'.format(i, lambda_, val_error))
```

```python
# Plot the traversal of the lambda/error space
plt.plot(sec_322_lambdas, sec_322_validation_errors, 'o-')
plt.show()
```

```python
# At this point we have iterated several times and the lambda values seem to be jumping all
# over the place. The best error value we managed to obtain was at iteration 14.
sec_322_OPTIMAL_ITERATION = 14
sec_322_LAMBDA = sec_322_lambdas[sec_322_OPTIMAL_ITERATION]
sec_322_OPTIMAL_ERROR = sec_322_validation_errors[sec_322_OPTIMAL_ITERATION]

print('Final LAMBDA = {:4f} with a validation error of {:4f}'.format(sec_322_LAMBDA,
                                                                     sec_322_OPTIMAL_ERROR))
print('This is an improvement of {:4f} from the original validation error with the naive lambda.'.format(
    sec_322_validation_error_by_rank[sec_322_RANK] - sec_322_OPTIMAL_ERROR))
print('And an improvement of {:4f} from the error in the baseline recommender system.'.format(
    0.867 - sec_322_OPTIMAL_ERROR))
```

```python
# Continuous colors
from matplotlib.colors import rgb2hex 
cmap = plt.get_cmap('viridis')

def to_red_green(val):
    val = val * 100
    red = 255 if val<50 else round(256 - (val-50)*5.12)
    green = 255 if val>50 else round((val)*5.12)
    return rgb2hex([red/255, green/255, 0])

source = ColumnDataSource(
    data={
        "x": sec_322_ex310_product_feature_x,
        "y": sec_322_ex310_product_feature_y,
        "name": [x[1] for x in sec_322_selected_movies],
        "color": [to_red_green(x[2]) for x in sec_322_selected_movies],
        "genres": [sec_322_movieId_to_genres[x[0]] for x in sec_322_selected_movies]
    })

hover = HoverTool(
    tooltips=[
        ("Name", "@name"),
        ("Genres", "@genres"),
    ])
tools = [hover, ResetTool(), PanTool(), WheelZoomTool(), SaveTool()]

p = figure(plot_width=720, plot_height=480, tools=tools, title="Mouse over the dots")
p.circle("x", "y", source=source, size=15, color="color", alpha=0.8)
show(p, notebook_handle=True)
```
